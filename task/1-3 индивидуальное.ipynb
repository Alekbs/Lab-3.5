### Индивидуальное задание к лабораторной роботе 3.5

### Условие:
##### Даны данные финансовых показателей АО "Ставропольский инструментальный завод" за 2012 - 2020 г.
##### Найти темпы роста выручки: изменение выручки от года к году в процентном соотношении. Отобразить отношение дохода к чистой прибыли. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Исходные данные:

```python
chartData = {
    "2012": {"assets": 119152000, "revenue": 65230000, "net_profit": -216000, "net_assets": 58996000},
    "2013": {"assets": 109338000, "revenue": 65615000, "net_profit": 2482000, "net_assets": 61756000},
    "2014": {"assets": 117544000, "revenue": 61689000, "net_profit": 3371000, "net_assets": 65094000},
    "2015": {"assets": 131421000, "revenue": 80677000, "net_profit": 9313000, "net_assets": 74374000},
    "2016": {"assets": 152916000, "revenue": 94900000, "net_profit": 18071000, "net_assets": 92411000},
    "2017": {"assets": 169103000, "revenue": 98496000, "net_profit": 14435000, "net_assets": 106813000},
    "2018": {"assets": 187930000, "revenue": 91006000, "net_profit": 13505000, "net_assets": 120381000},
    "2019": {"assets": 202206000, "revenue": 86085000, "net_profit": 9051000, "net_assets": 129399000},
    "2020": {"assets": 216863000, "revenue": 68615000, "net_profit": 10596000, "net_assets": 139962000}   
}
```

## 1)Линейный график



Чтобы решить задачу, нужно сначала вычислить себестоимость производства от 1 до 10 тонн картофеля на первом участке и оптовую выручку от их продажи, а затем вычислить разницу между ними - это и будет дифференциальной рентой.


```python
years = list(chartData.keys())
revenue = [chartData[year]["revenue"] for year in years]
revenue_growth_rates = [(revenue[i] - revenue[i+1]) / revenue[i+1] * 100 for i in range(len(revenue)-1)]
```

```python
# Строим график
plt.plot( years[:-1], revenue_growth_rates, marker='o', label='Себестоимость')
plt.title("Темпы роста выручки по годам")
plt.xlabel("Год")
plt.ylabel("Выручка")
plt.show()
```

## 2)Столбчатая диаграмма 

Получим дифференциальнаю ренту анналогично вычислениям в первом пункте 

```python
net_profit = [chartData[year]["net_profit"] for year in years]

# Определение ширины столбцов и смещение по оси x
bar_width = 0.35
index = np.arange(len(years))

# Построение столбцов
plt.bar(index, revenue, bar_width, label="Доход")
plt.bar(index + bar_width, net_profit, bar_width, label="Чистая прибыль")

plt.title("Доход и Чистая прибыль по годам")
plt.xlabel("Год")
plt.ylabel("Сумма")
plt.xticks(index + bar_width/2, years)  # Установка подписей оси x
plt.legend()

plt.show()
```

## 3)Круговая диаграмма

```python
net_profit = [chartData[year]["net_profit"] for year in years[1:]]

# Построение круговой диаграммы
plt.pie(net_profit, labels=years[1:], autopct='%1.1f%%')

plt.title("Чистая прибыль по годам")
plt.axis('equal')

plt.show()
```
